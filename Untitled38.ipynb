Importing the required libraries

```python
from pyforest import*
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
stop_words = set(stopwords.words('english'))
from wordcloud import WordCloud 
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
import warnings
warnings.filterwarnings('ignore')
```

```python
df = df.drop(['Unnamed: 0','img_source', 'video_link'],axis=1)
```

```python
df.head()
```

```python
for col in ['title','categories','tags']:
    df[col] = df[col].str.replace('\d+','')
    df[col] = df[col].str.replace(r'[_,-]+',' ')
    
```

```python
df.head()
```

Most viewed adult content

```python
mvac = df[['nb_views','title']].sort_values(by='nb_views', ascending=False).head(6)
mvac.sort_values(by='nb_views',ascending=True, inplace=True)
plt.figure(figsize=(15,10))
ax=sns.barplot(
    y=mvac['title'], 
    x=mvac['nb_views']
)
plt.xticks(rotation= 45)
plt.ylabel('Video Titles')
plt.xlabel('Number of views')
plt.title('Most viewed Adult videos')
```

Most viewed adult category

```python
stopwords = set(stop_words)
wordcloud = WordCloud(
                         #background_color='gray',
                         stopwords=stopwords,
                         max_words=100,
                         max_font_size=200, 
                         width=1000, height=600,
                         random_state=69,
                         ).generate(" ".join(df['tags'].astype(str)))

fig = plt.figure(figsize = (12,12), facecolor=None)
plt.imshow(wordcloud)
plt.title("Frequently Used Words In Tags", fontsize=18)
plt.axis('off')
plt.show()
```
